## Concise version of softmax(use framework)

才意识到torch的名字被改成d2l了lol，但是确实是d2l module里的，这样命名没问题。d2l是python的一个官方库，自己pip下，然后torch是里面一个子库。


```python
import torch
from torch import nn
from torch.nn import functional as F
from d2l import torch as d2l
```

```python
class SoftmaxRegression(d2l.Classifier): #@save
    def __init__(self, num_outputs, lr):
        super().__init__()
        self.save_hyperparameters()
        self.net = nn.Sequential(
                    nn.Flatten(), 
                    nn.LazyLinear(), 
                    )
        
    def forward(self, X): 
        '''因为用了Flatten, 直接被压成batch_size * -1'''
        return self.net(X)
```
