## Softmax 



$\mathbf{o} = \mathbf{W} \mathbf{x} + \mathbf{b}$
linear的问题是我们不能吧o的值框定在：
1. 0 < o < 1

然鹅，o有可能是负的，有可能超过1

为了ensure nonnegativity, 我们引入exp, 再normalize一下。

最后得到：$\hat{\mathbf{y}} = \mathrm{softmax}(\mathbf{o}) \quad \textrm{where}\quad \hat{y}_i = \frac{\exp(o_i)}{\sum_j \exp(o_j)}.$
